```python
import numpy as np
import pandas as pd
```

```python
df_eqtl = pd.read_csv('FUMA_dowloads_all_tissue/eQTL/eqtl_ENDOMETRIOSIS.txt', sep='\t')
df_snps = pd.read_csv('FUMA_dowloads_all_tissue/SNPs/snps_ENDOMETRIOSIS.txt', sep='\t')
```

```python
relevant_snps = df_snps[df_snps['uniqID'].isin(df_eqtl['uniqID'].unique())]
dict1 = dict(zip(relevant_snps['uniqID'], relevant_snps['rsID']))
```

```python
df_eqtl['rsID'] = df_eqtl['uniqID'].map(dict1)
```
